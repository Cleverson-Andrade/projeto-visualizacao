```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

# =========================================================================
# LINK DO MEU REPOSITÓRIO OFICIAL (DATASET DO PROJETO)
# =========================================================================
LINK_DO_SEU_DATASET = "https://github.com/Cleverson-Andrade/projeto-visualizacao"

# =========================================================================
# 1. CONSTRUÇÃO E PROCESSAMENTO DAS MINHAS VISUALIZAÇÕES
# =========================================================================
sns.set_theme(style="whitegrid")
plt.rcParams['text.color'] = '#333333'
plt.rcParams['axes.labelcolor'] = '#333333'

# Minha estrutura de dados integrando câmbio e safras agrícolas
dados_agricolas = {
    'Ano': [2021, 2022, 2023, 2024, 2025],
    'Produtividade_Sacas_ha': [85, 92, 78, 95, 102],
    'Preco_Saca_BRL': [92.50, 84.00, 55.30, 68.00, 72.00],
    'Dolar_Medio': [5.39, 5.16, 4.99, 5.15, 5.45]
}
df_projeto = pd.DataFrame(dados_agricolas)

# Meu Plot 1 – Linha Contínua
plt.figure(figsize=(9, 4))
sns.lineplot(data=df_projeto, x='Ano', y='Preco_Saca_BRL', marker='o', color='darkgreen', linewidth=2.5)
plt.title('Evolução do Preço Médio da Saca de Milho (2021 - 2025)', fontsize=12, fontweight='bold', pad=10)
plt.xlabel('Ano Comercial', fontsize=10)
plt.ylabel('Preço por Saca (R$)', fontsize=10)
plt.xticks(df_projeto['Ano'])
plt.tight_layout()
plt.savefig('grafico1_linha.png', dpi=300)
plt.show()  # Mostra o gráfico na tela antes de encerrar
plt.close()

# Meu Plot 2 – Barras Discretas
plt.figure(figsize=(9, 4))
sns.barplot(data=df_projeto, x='Ano', y='Produtividade_Sacas_ha', palette='Greens_d')
plt.title('Rendimento Médio de Produtividade por Safra', fontsize=12, fontweight='bold', pad=10)
plt.xlabel('Ano de Colheita', fontsize=10)
plt.ylabel('Produtividade Média (Sacas / ha)', fontsize=10)
plt.tight_layout()
plt.savefig('grafico2_barras.png', dpi=300)
plt.show()  # Mostra o gráfico na tela antes de encerrar
plt.close()

# Meu Plot 3 – Regressão Estatística / Dispersão
plt.figure(figsize=(9, 4))
sns.regplot(data=df_projeto, x='Dolar_Medio', y='Preco_Saca_BRL',
            scatter_kws={'s':80, 'color':'teal'}, line_kws={'color':'orange', 'linewidth':2})
plt.title('Correlação: Taxa de Câmbio (USD/BRL) vs Preço da Saca de Milho', fontsize=12, fontweight='bold', pad=10)
plt.xlabel('Câmbio Médio Anual (USD/BRL)', fontsize=10)
plt.ylabel('Preço da Saca (R$)', fontsize=10)
plt.tight_layout()
plt.savefig('grafico3_dispersao.png', dpi=300)
plt.show()  # Mostra o gráfico na tela antes de encerrar
plt.close()

# =========================================================================
# 2. COMPILAÇÃO NATIVA DO MEU RELATÓRIO PDF (MODELO EXATO INSTITUCIONAL)
# =========================================================================
def gerar_pdf_modelo():
    doc = SimpleDocTemplate("modelo_projeto_final.pdf", pagesize=letter, rightMargin=40, leftMargin=40, topMargin=40, bottomMargin=40)
    story = []
    styles = getSampleStyleSheet()

    text_style = ParagraphStyle('ModelText', parent=styles['Normal'], fontName='Helvetica', fontSize=11, leading=16, textColor=colors.HexColor('#222222'))
    bold_style = ParagraphStyle('ModelTextBold', parent=text_style, fontName='Helvetica-Bold')
    desc_style = ParagraphStyle('ModelDesc', parent=text_style, fontName='Helvetica-Oblique', fontSize=9.5, textColor=colors.HexColor('#555555'), spaceBefore=4, spaceAfter=15)

    # Dados acadêmicos estruturados
    story.append(Paragraph("<b>Nome:</b> Cleverson Moura Andrade", text_style))
    story.append(Paragraph("<b>RGM:</b> 45100390", text_style))
    story.append(Paragraph("<b>Instituição:</b> Cruzeiro do Sul Virtual", text_style))
    story.append(Paragraph("<b>Curso:</b> Ciência de Dados", text_style))
    story.append(Paragraph(f"<b>Link do dataset utilizado:</b> <font color='#0066CC'><u>{LINK_DO_SEU_DATASET}</u></font>", text_style))
    story.append(Spacer(1, 15))

    # Plot 1
    story.append(Paragraph("<b>Plot 1 – Análise Temporal de Preços (Unidade: R$)</b>", bold_style))
    story.append(Spacer(1, 5))
    story.append(Image("grafico1_linha.png", width=440, height=195))
    story.append(Paragraph("<i>Descrição: Gráfico de linhas contínuas que utilizei por ser ideal para o monitoramento de tendências históricas e flutuações de preços da commodity ao longo das safras.</i>", desc_style))

    # Plot 2
    story.append(Paragraph("<b>Plot 2 – Rendimento Físico por Safra (Unidade: Sacas / ha)</b>", bold_style))
    story.append(Spacer(1, 5))
    story.append(Image("grafico2_barras.png", width=440, height=195))
    story.append(Paragraph("<i>Descrição: Gráfico de barras verticais discretas que apliquei para permitir a comparação direta e imediata do volume total de produtividade entre os períodos fechados.</i>", desc_style))

    # Plot 3
    story.append(Paragraph("<b>Plot 3 – Correlação Macroeconômica (Unidade: Proporção USD/BRL)</b>", bold_style))
    story.append(Spacer(1, 5))
    story.append(Image("grafico3_dispersao.png", width=440, height=195))
    story.append(Paragraph("<i>Descrição: Gráfico de dispersão (Scatter Plot) com linha de tendência estatística que implementei para avaliar a dependência direta dos preços domésticos frente à taxa de câmbio.</i>", desc_style))

    doc.build(story)

if __name__ == '__main__':
    gerar_pdf_modelo()
```
